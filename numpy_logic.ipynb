```python
immport csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
months = np.array(["January", "February", "March", "April", "May", "June","July", "August", "September", "October","November", "December"])
sales = []
print("Enter the sales of each month:")
with open("Sales.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Month", "Sales"])
    for month in months:
        value = float(input(f"Enter the sales amount in the month of {month}: Rs "))
        sales.append(value)
        writer.writerow([month, value])
df = pd.read_csv("Sales.csv")
sales_np = df["Sales"].to_numpy()
#LINE GRAPH
plt.figure(figsize=(10, 5))
plt.plot(df["Month"], sales_np, marker='o', linewidth=2)
plt.xlabel("Month")
plt.ylabel("Sales (Rs)")
plt.title("Monthly Sales Trend", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
#BAR GRAPH
plt.figure(figsize=(10, 5))
bars = plt.bar(df["Month"], sales_np)
plt.xlabel("Month")
plt.ylabel("Sales (Rs)")
plt.title("Monthly Sales Comparison", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),f'{bar.get_height():.0f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
#HORIZONTAL BAR GRAPH
plt.figure(figsize=(10, 5))
bars = plt.barh(df["Month"], sales_np)
plt.xlabel("Sales (Rs)")
plt.ylabel("Month")
plt.title("Monthly Sales (Horizontal View)", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,f'{bar.get_width():.0f}', va='center')
plt.tight_layout()
plt.show()
#PIE CHART
plt.figure(figsize=(7, 7))
plt.pie(
    sales_np,
    labels=df["Month"],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Month-wise Sales Distribution", fontsize=14)
plt.tight_layout()
plt.show()
print("\n--- SALES RECORD OF THE COMPANY ---")
print("Total Sales: Rs.", np.sum(sales_np))
print("Average Sales: Rs.", np.mean(sales_np))
print("Highest Sales: Rs.", np.max(sales_np))
print("Lowest Sales: Rs.", np.min(sales_np))
print("Highest Sales Month:", df.loc[sales_np.argmax(), "Month"])
print("Lowest Sales Month:", df.loc[sales_np.argmin(), "Month"])
print("Months with sales above average:",df["Month"][sales_np > np.mean(sales_np)].to_list())
print("Months with sales below average:",df["Month"][sales_np < np.mean(sales_np)].to_list())
```
